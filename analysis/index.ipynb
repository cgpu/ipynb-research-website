## Tutorial overview

The purpose of this tutorial is to gain some familiarity with literate programming with Jupyter Notebooks. 

All the results and plots presented in the pages below should be reproducable on your own environment. Follow the
[dependencies instructions](dependencies/README.md) to install the required dependencies before starting with the individual sections of the tutorial. 

Here are the rendered versions of the Jupyter Notebooks that can be found in the [analysis folder](analysis). We will work through each section during the tutorial. The html files were generated by executing the [Jupyter notebooks](https://github.com/cgpu/ipynb-research-website/tree/ef1dc1a2f283ea873212ee5c3f25e7bffd3566b1/analysis).
